# Line Plot

```python
import matplotlib
import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
#create figure and axis
fig=plt.figure()
ax=plt.axes()
#declare a variable
x=np.linspace(0,10,1000)
#plot the sin function
ax.plot(x,np.sin(x),'b-')
```

# Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color = 'blue');
```

# Histogram

```python
#The plt.hist() function can be used to plot a simple histogram as follows:
data1 = np.random.randn(10000)
plt.hist(data1);
```

# Bar Chart

```python
data = [5. , 90. , 50. , 20.,89.,30.,50.,34.]
plt.bar(range(len(data)), data)
plt.show() 
```

# Horizontal Bar Chart

```python
data = [5. , 90. , 50. , 20.,89.,30.,50.,34.]
plt.barh(range(len(data)), data)
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(7,7))
x = [35, 25, 20, 20]
labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
plt.pie(x, labels=labels);
plt.show()
```

# Adding a grid

```python
x=np.arange(1,5)
plt.plot(x*2,x**3,x/3)
plt.grid(True)
plt.show()
```

```python
#LIMIT
x15 = np.arange(1, 5)
plt.plot(x15, x15*1.5, x15, x15*3.0, x15, x15/3.0)
plt.xlim([1.0, 4.0])
plt.ylim([0.0, 12.0])
```

```python
#Adding lavel
plt.plot([1, 3, 2, 4])
plt.xlabel('This is the X axis')
plt.ylabel('This is the Y axis')
plt.show()
```

```python
#Adding a title
plt.plot([1, 3, 2, 4])
plt.title('First Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```
